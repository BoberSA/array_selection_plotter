```python
import numpy as np

# подключение модуля array_selection.py, импорт класса selection_plotter
from array_selection import selection_plotter as sp
```

```python
# размер шрифта для визуализации массивов
sp.fontsize = 12
```

```python
# исходный массив (можно использовать и одномерный массив)
arr = np.arange(36).reshape(6, 6)

# представление массива, визуализирующее любой выбор
v = arr.view(type=sp)
```

```python
v # без квадратный скобок визуализации не будет
```

```python
v[1, 2] # выбор одного элемента
```

```python
v[1] # выбор строки с индексом 1
```

```python
v[::2] # выбор строк при помощи среза (с шагом 2)
```

```python
v[2:5, -4:-1] # выбор нескольких строк и столбцов срезами
```

```python
# построение маски по исходному массиву, чтобы не было лишней визуализации
mask = arr[1] % 3 == 0
mask
```

```python
v[:, mask] # выбор столбцов по маске
```

```python
mask2 = arr % 7 == 0 # двумерная маска
```

```python
v[mask2] # выбор по двумерной маске
```

```python
# выбор элементов с индексами [1,2], [3,4], [4,5]
v[[1, 3, 4], [2, 4, 5]]
```

```python
# выбор элементов на пересечении строк [1, 3, 5] и столбцов [0, 2, 4]
rows = np.array([1, 3, 5]).reshape(-1, 1)
cols = np.array([0, 2, 4])
rows, cols
```

```python
v[rows, cols]
```
